## CONSULTAS SQLDF

CONSULTA 1

```r
consulta<-sqldf( "SELECT * FROM publicidad ")
```

```r
consulta
```

CONSULTA 2

```r
consulta2<-sqldf("SELECT idPublicidad AS ID, ENTIDAD , VC_RUC_ENTIDAD AS RUC, VC_PUBLICIDAD_ANNO AS AÑO
                 FROM publicidad
                 WHERE (VC_PUBLICIDAD_PROVEEDOR='PRODUCCIONES GENESIS S.A.C.')")
```

```r
consulta2
```

 CONSULTA 3

```r
Consulta3<-sqldf( "SELECT idPublicidad AS ID,ENTIDAD ,FUENTE
                      FROM publicidad
                       WHERE  FUENTE = (SELECT FUENTE
                      FROM publicidad 
                     WHERE FUENTE ='CANON, SOBRECANON, REGALIAS Y PARTICIPACIONES')")
```

```r
Consulta3
```

CONSULTA 4

```r
consulta4<-sqldf("SELECT idPublicidad AS ID,ENTIDAD,VC_PUBLICIDAD_MES ,COUNT(VC_PUBLICIDAD_MES) AS CANTIDAD_DE_PUBLICIDADES_MES
                    FROM publicidad
                    WHERE  ENTIDAD IN (SELECT  ENTIDAD
                                      FROM publicidad
                                      WHERE ENTIDAD='CONGRESO DE LA REPÚBLICA (CONGRESO)')
                                      GROUP BY VC_PUBLICIDAD_MES  " )
```

```r
consulta4
```
